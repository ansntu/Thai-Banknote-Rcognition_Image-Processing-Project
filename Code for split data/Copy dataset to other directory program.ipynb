```python
import glob
import numpy as np
from sklearn.model_selection import train_test_split
import shutil

import os

file_urls = glob.glob("../Dataset/*/*")
print("Total File found:", len(file_urls))

labels = list(map(lambda x: (x.split('/'))[-2], file_urls))
# print(labels)

train_size = 0.8
val_size = 0.125

train_val_X, test_X, train_val_y, test_y = train_test_split(file_urls, labels, train_size=train_size, random_state=70, stratify=labels)
train_X, val_X, train_y, val_y = train_test_split(train_val_X, train_val_y, test_size=val_size, random_state=70, stratify=train_val_y)

print("Total Files after splitting:", len(train_X) + len(val_X) + len(test_X))

print(len(val_X))
print(len(test_X))
dataset_dict = {
    "train": {
        "X": train_X,
        "y": train_y
    },
    "val": {
        "X": val_X,
        "y": val_y
    },
    "test": {
        "X": test_X,
        "y": test_y
    }
}
# print(dataset_dict["train"]["X"])
```

```python
for key in dataset_dict:
    target_path = '../dataset_splits/' + key
    print(target_path)
    
    current_labels = dataset_dict[key]["y"]
    count = 0
    for x in dataset_dict[key]["X"]:
        
        cpy_target_path = target_path + '/' + current_labels[count]
        os.makedirs(cpy_target_path, exist_ok=True)
        
        shutil.copy(x, cpy_target_path)
        print(x, current_labels[count], cpy_target_path)
        count += 1
```
